```python
import pandas as pd

# Load both datasets
jan = pd.read_csv('../data/bengaluru-meteorology-summary-jan-2015.csv')
jul = pd.read_csv('../data/bengaluru-meteorology-summary-jul-2015.csv')

# Combine them into one DataFrame
weather_data = pd.concat([jan, jul], ignore_index=True)

# Show first few rows and info
display(weather_data.head())
weather_data.info()

# Save the combined file
weather_data.to_csv('../data/bangalore_weather.csv', index=False)
print("✅ Combined and saved as bangalore_weather.csv")
```

```python
import pandas as pd

# Reload the combined weather file
weather_data = pd.read_csv('../data/bangalore_weather.csv', header=None)

# Assign proper column names (based on IMD data structure)
weather_data.columns = [
    "Year", "Month", "Day", "Hour",
    "Rainfall_mm", "SolarRadiation_Wm2",
    "Temperature_C", "WindSpeed_mps", "WindDirection_deg"
]

# Show first few rows
display(weather_data.head())

# Save the cleaned version
weather_data.to_csv('../data/bangalore_weather_clean.csv', index=False)
print("✅ Cleaned and saved as bangalore_weather_clean.csv")
```
