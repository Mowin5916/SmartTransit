```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('predictions.csv')
print(df.head())
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='Live_Congestion',
    y='Predicted_Passenger_Count',
    hue='Delay_Minutes',
    palette='coolwarm',
    s=80
)
plt.title("🚏 Passenger Count vs Live Congestion (Color = Delay Minutes)", fontsize=13)
plt.xlabel("Live Congestion (%)")
plt.ylabel("Predicted Passenger Count")
plt.legend(title="Delay (min)")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
import numpy as np

# add realistic fluctuations
np.random.seed(42)
df['Live_Congestion'] = 50 + np.random.randint(20, 80, size=len(df)) / 2  # 50–90%
df['Delay_Minutes'] = np.random.randint(0, 25, size=len(df))  # 0–25 mins

df.to_csv("predictions_updated.csv", index=False)
print("✅ Updated with realistic traffic variations & saved as predictions_updated.csv")
```

```python
# Paste this into a new Jupyter cell and run
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# 1) Load
df = pd.read_csv("predictions_updated.csv")

# 2) Quick stats & correlation
print("Rows:", len(df))
print("\nNumeric summary:")
print(df[['Predicted_Passenger_Count','Live_Congestion','Delay_Minutes']].describe().round(2).to_string())

corr = df[['Predicted_Passenger_Count','Live_Congestion','Delay_Minutes']].corr()
print("\nCorrelation matrix:")
print(corr.round(3))

# 3) Scatter plot: congestion vs predicted passengers (color = delay)
plt.figure(figsize=(10,6))
sc = sns.scatterplot(
    data=df,
    x='Live_Congestion',
    y='Predicted_Passenger_Count',
    hue='Delay_Minutes',
    palette='coolwarm',
    size='Delay_Minutes',
    sizes=(20,200),
    alpha=0.85,
    edgecolor='w'
)
plt.title("Passenger Count vs Live Congestion (color = Delay minutes)")
plt.xlabel("Live Congestion (%)")
plt.ylabel("Predicted Passenger Count")
plt.legend(title="Delay (min)", bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 4) Alerts: heavy traffic or big delays
# Thresholds you can adjust: congestion >75% or delay > 15 minutes
alerts = df[(df['Live_Congestion'] > 75) | (df['Delay_Minutes'] > 15)].copy()
alerts['Alert'] = "⚠️ Heavy traffic or long delay"
alerts.to_csv("alerts_updated.csv", index=False)
print(f"\n✅ Alerts generated: {len(alerts)} saved to alerts_updated.csv")

if len(alerts) > 0:
    display(alerts[['Date','Route_Name','Time_Slot','Live_Congestion','Delay_Minutes','Predicted_Passenger_Count','Alert']].head(10))

# 5) Top 10 busiest routes by average predicted passengers
top_routes = df.groupby('Route_Name')['Predicted_Passenger_Count'].mean().sort_values(ascending=False).head(10)
print("\nTop 10 routes (avg predicted passengers):")
print(top_routes.round(1))

# Bar plot for top routes
plt.figure(figsize=(10,5))
sns.barplot(x=top_routes.values, y=top_routes.index)
plt.xlabel("Average Predicted Passenger Count")
plt.ylabel("Route Name")
plt.title("Top 10 Routes by Avg Predicted Passenger Count")
plt.tight_layout()
plt.show()
```
