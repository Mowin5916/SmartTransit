```python
import pandas as pd

# Load merged dataset (corrected to CSV)
data = pd.read_csv('merged_transit_data.csv')

# Display info
print("✅ Loaded merged_transit_data.csv successfully!")
print("Rows:", data.shape[0], " | Columns:", data.shape[1])
print("\nColumns:", list(data.columns))
data.head()
```

```python
# Make a copy to avoid overwriting
df = data.copy()

# Step 1: Drop duplicate or unnecessary columns
df.drop(columns=['Year', 'Month', 'Day', 'Hour', 
                 'Rainfall_mm_y', 'SolarRadiation_Wm2', 
                 'Temperature_C_y', 'WindSpeed_mps', 
                 'WindDirection_deg'], inplace=True, errors='ignore')

# Step 2: Rename the main columns for clarity
df.rename(columns={
    'Temperature_C_x': 'Temperature_C',
    'Rainfall_mm_x': 'Rainfall_mm',
    'Live_Congestion_Level(%)': 'Live_Congestion',
    'Usual_Congestion_Level(%)': 'Usual_Congestion',
    'Delay_Minutes_per_10km': 'Delay_Minutes'
}, inplace=True)

# Step 3: Fill missing numeric values with median, categorical with mode
for col in ['Temperature_C', 'Rainfall_mm', 'Live_Congestion', 'Usual_Congestion', 'Live_Speed_kmph', 'Delay_Minutes']:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)

for col in ['Weather_Condition', 'Accident_Risk_Level']:
    if col in df.columns:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Step 4: Show final columns and few sample rows
print("✅ Cleaned dataset ready!")
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
print("Columns:", list(df.columns))
df.head()
```

```python
import numpy as np

# Create a clean copy
df = data.copy()

# Drop unnecessary columns safely
df = df.drop(columns=[
    'Year', 'Month', 'Day', 'Hour',
    'Rainfall_mm_y', 'SolarRadiation_Wm2',
    'Temperature_C_y', 'WindSpeed_mps',
    'WindDirection_deg'
], errors='ignore')

# Rename key columns for clarity
df = df.rename(columns={
    'Temperature_C_x': 'Temperature_C',
    'Rainfall_mm_x': 'Rainfall_mm',
    'Live_Congestion_Level(%)': 'Live_Congestion',
    'Usual_Congestion_Level(%)': 'Usual_Congestion',
    'Delay_Minutes_per_10km': 'Delay_Minutes'
})

# Fill numeric columns with median safely
numeric_cols = ['Temperature_C', 'Rainfall_mm', 'Live_Congestion',
                'Usual_Congestion', 'Live_Speed_kmph', 'Delay_Minutes']

for col in numeric_cols:
    if col in df.columns and df[col].notna().any():
        df[col] = df[col].fillna(df[col].median())

# Fill categorical columns safely
for col in ['Weather_Condition', 'Accident_Risk_Level']:
    if col in df.columns and df[col].notna().any():
        df[col] = df[col].fillna(df[col].mode().iloc[0])
    else:
        df[col] = df[col].fillna("Unknown")

# Show summary
print("✅ Cleaned dataset ready (fixed warnings & errors)!")
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
print("Columns:", list(df.columns))
df.head()

```
