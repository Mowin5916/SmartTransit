```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

print("✅ Environment ready — pandas and sklearn imported successfully!")
```

```python
# Load the cleaned dataset
df = pd.read_csv('../data/merged_transit_data.csv')

print("✅ Loaded dataset successfully!")
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
df.head()
```

```python
# Make a copy
encoded_df = df.copy()

# Define categorical columns
cat_cols = ['Route_ID', 'Route_Name', 'Time_Slot',
            'Weather_Condition', 'Accident_Risk_Level']

# Label encoding
encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))
    encoders[col] = le

# Fill any remaining missing numeric values
num_cols = ['Live_Congestion', 'Usual_Congestion',
            'Live_Speed_kmph', 'Delay_Minutes']
for col in num_cols:
    if encoded_df[col].isna().any():
        encoded_df[col].fillna(encoded_df[col].median(), inplace=True)

print("✅ Categorical encoding complete!")
encoded_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Make a copy
encoded_df = df.copy()

# Handle both naming possibilities (old/new)
col_map = {
    'Live_Congestion_Level(%)': 'Live_Congestion',
    'Usual_Congestion_Level(%)': 'Usual_Congestion',
    'Delay_Minutes_per_10km': 'Delay_Minutes'
}
encoded_df.rename(columns=col_map, inplace=True)

# Define categorical columns
cat_cols = ['Route_ID', 'Route_Name', 'Time_Slot',
            'Weather_Condition', 'Accident_Risk_Level']

# Encode categorical data
encoders = {}
for col in cat_cols:
    if col in encoded_df.columns:
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))
        encoders[col] = le

# Fill any missing numeric values
num_cols = ['Live_Congestion', 'Usual_Congestion',
            'Live_Speed_kmph', 'Delay_Minutes']
for col in num_cols:
    if col in encoded_df.columns:
        encoded_df[col].fillna(encoded_df[col].median(), inplace=True)

print("✅ Encoding & renaming successful!")
print("Columns:", encoded_df.columns.tolist())
encoded_df.head()
```
